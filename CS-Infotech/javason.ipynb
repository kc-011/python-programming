## JSON

### Serialization
Python Obj -----> JSON

```python
import json

emp = {"name":"shalini", "age":20, "projects":["jdsjd","skdjks","ksdks"]}
jstr = json.dumps(emp)
print(type(jstr))
print(jstr)
```

### Deserialization
JSON -----> Python Obj

```python
jstr = '[{"name": "shalini", "age": 20, "projects": ["jdsjd", "skdjks", "ksdks"]}, {"name": "shabaz", "age": 25, "projects": ["fkjdf", "skdls", "lsmdsk"]}]'
print(type(jstr))
lst = json.loads(jstr)
for d in lst:
    print(d.get("name","NA"))
    print(d.get("projects","NA"))
```
