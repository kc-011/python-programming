## Q. Generate a strong password 
Necessary characters - Uppercase, Lowercase, Digit

```python
import string
import random
from itertools import chain

def genpwd(length=8):
    upper = string.ascii_uppercase
    lower = string.ascii_lowercase
    digit = string.digits
    special = string.punctuation
    lst = [random.choice(upper), random.choice(lower), random.choice(digit)]
    while len(lst)!=length:
        lst.append(random.choice(list(chain(upper,lower,digit,special))))
    random.shuffle(lst)
    pwd = ''.join(lst)
    print(f"Your password is {pwd}")

genpwd(12)
```

## Q. Check if a password is strong or weak
Should return a bool and a list of errors (what stuff is missing in the pwd) <br>
Pwd length, uppercase, digit and special chars

```python
def check_pwd(pwd):
    import string
    i=0
    errors_lst = []
    if len(pwd) < 8:
        errors_lst.append('Password length < 8 Chars')
    if not any([ch.isupper() for ch in pwd]):
        errors_lst.append('Password has no uppercase letter')
    if not any(ch.isdigit() for ch in pwd):
        errors_lst.append('Password has no digit')
    for ch in pwd:
        if ch in string.punctuation:
            i+=1
    if i<1:
        errors_lst.append('Password has no special characters')
    if len(errors_lst)!=0:
        for i in errors_lst:
            print('-->', i)
        return False
    else:
        return True, errors_lst
    
check_pwd('@!1lOllloo')
```

## Q. Make a function to print HTML tags

```python
def html_tag(tag, content, *attr):
    attributes = " ".join(attr)

    if attributes:
        return f"<{tag} {attributes}>{content}</{tag}>"
    else:
        return f"<{tag}>{content}</{tag}>"

html_tag('a', 'Visit now!', "href='www.google.com'", "target='_blank'")
# html_tag('a', 'Yooo')
```

## Q. Make a function to pass an SQL query

```python
def sql_query(**condition):
    query = f"SELECT {condition['select']}\nFROM {condition['table']}\n"
    if condition.get('where'):
        query += f"WHERE {condition['where']}\n"
    if condition.get('order_by'):
        query += f"ORDER BY {condition['order_by']}\n"
    if condition.get('limit'):
        query+= f"LIMIT {condition['limit']}\n"

    print(query)
    

sql_query(select='students', table='ICSE', where='marks >= 90', order_by='class')
```
