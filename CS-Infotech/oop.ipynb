### Class Methods (decorator)

```python
class Student:
    no_student = 0 # class variable
    
    def __init__(self, name, standard):
        self.name = name
        self.std = standard
        Student.no_student += 1

    def __str__(self):
        return f"{self.name} is in Class {self.std}"
        
    @classmethod #to access class variable
    def showcount(cls):
        return cls.no_student

s1 = Student("Karan", 12)
s2 = Student("Ankush", 9)
print(s1)
Student.showcount()
```

### What if we don't use the decorator??

```python
class Student:
    no_student = 0 # class variable
    
    def __init__(self, name, standard):
        self.name = name
        self.std = standard
        Student.no_student += 1

    def __str__(self):
        return f"{self.name} is in Class {self.std}"
        
    def showcount(self):
        return Student.no_student
        # or return self.no_student

s1 = Student("Karan", 12)
s2 = Student("Ankush", 9)
print(s1)
s1.showcount()
```

It works, but to access a class variable, we have to call it from an object, which does not make sense as the class variable should be directly called from the class itself and not any object as that variable is the same for all objects. Hence it works but is not Python-friendly code.

### Magic method: Add

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"({self.x},{self.y})"
    
    def __add__(self, other):
        if isinstance(other, Point):
            pt = Point(self.x + other.x, self.y + other.y)
            return pt
        elif isinstance(other, int):
            pt = Point(self.x + other, self.y + other)
            return pt
        else:
            return None

p = Point(5,3)
p1 = Point(7,4)
p2 = p+3
p3 = p1 + 9
p4 = p+p1
p5 = p3 +"idkdmfkd"
print(p2) # int
print(p3) # int
print(p4) # obj
print(p5) # None
```

### Magic Methods: Subtract, Equal

```python
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __sub__(self, other):
        if isinstance(other, Point):
            return round(math.sqrt((self.x - other.x)**2+(self.y - other.y)**2),2)
        else:
            raise Exception("Argument must be of Point Class")
        
    def __eq__(self, other):
        if self.x == other.x and self.y == other.y:
            return True
        else:
            return False

p1 = Point(3,2)
p2 = Point(5,6)
p3 = p1-p2
p4 = Point(3,2)
print(p3)
print(p1==p2)
print(p1==p4)
```

## Property Decorator

```python
class User:
    def __init__(self, unam, upwd):
        self.__unam = unam
        self.__upwd = upwd
    
    @property
    def userpwd(self):
        return "*"*len(self.__upwd)
    
    @userpwd.setter
    def userpwd(self, value):
        if 6<=len(value)<12:
            self.__upwd = value
        else:
            raise Exception("Password must be between 6 and 12 characters.")

    @property
    def username(self):
        return self.__unam
    
    @username.setter
    def username(self, value):
        if value=="":
            raise Exception("Username cannot be empty")
        else:
            self.__unam = value.lower()
    
u, p = input("Enter username, password: ").split(",")
ut = User(u,p)
print(ut.username,"        ", ut.userpwd)
```

## Inheritance

```python
class mystring(str):
    def __init__(self, value):
        # super().__init__()
        self.value = value
    
    def reverse(self):
        return self.value[::-1]
    
    def ispalindrome(self):
        return True if self.value == self.value[::-1] else False

m = mystring('shalini')
print(m.reverse())
print(m.ispalindrome())
print(m.capitalize(), m.upper())
```

In the above code, we inherited from 'str' class, which is why we can access all functions of 'str' class with the object outside the class (which is bad practice btw).

```python
class myintlist(list):
    def append(self, data):
        if isinstance(data, int):
            if data not in self:
                super().append(data)
            else:
                raise Exception('No duplicates allowed')
        else:
            raise Exception('Only integers allowed')
        
lst = myintlist()
lst.append(67)
lst.append(74)
print(lst)
```

In the above code, we have overriden the append func by making our own method, and when we need to access the actual append func from list obj, then we add <i>super()</i> before the func. <br>
Here also we are able to access all funcs of list obj outside of the class.

## Abstraction

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def calc_area(self):
        pass

class Circle(Shape):
    def __init__(self,r):
        self.r = r
    
    def calc_area(self):
        return f"Area of circle is {3.14*self.r*self.r}"
    
class Rect(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b
    
    def calc_area(self):
        return f"Area of rectangle is {self.l*self.b}"

c1 = Circle(4)
print(c1.calc_area())
r1 = Rect(9,5)
print(r1.calc_area())
```

### More practice on property decorator

```python
class Name:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, value):
        if value == "":
            raise ValueError("Name cannot be empty")
        else:
            self.__name = value
    
    @property
    def marks(self):
        if self.__marks >= 90:
            return 'A'
        elif self.__marks >=80:
            return 'B'
        elif self.__marks >=70:
            return 'C'
        else:
            return 'You flopped 😂'

    @marks.setter
    def marks(self, value):
        if value < 0 or value >100:
            raise Exception("Marks must between 0 and 100")
        else:
            self.__marks = value 
            
# All getters and setters must return private attributes otherwise the method will be called recursively

s = Name("Chawla", 91)
print(s.name)
s.name = "Karan"
print(s.name)
print(s.marks)
# s1 = Name("Amit", 105) # raises my exception
# s2 = Name("", 82) # raises my exception
s3 = Name("Ryan", 40)
print(s3.marks)
```

A property is used when an object should expose something that behaves like an attribute, while allowing validation, computation, or controlled access behind the scenes. Normal methods are better when the object is performing an action or operation.
