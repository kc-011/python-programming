# Lambda Function Questions
## Using map, filter, sorted, min/max

```python
res = lambda x: x**2
res(5)
```

```python
res = lambda x,y:x+y
res(9,5)
```

```python
res = lambda x,y: x if x>y else y
res(26,92)
```

```python
res = lambda x: 'Even' if x%2==0 else 'Odd'
res(67)
```

```python
res = lambda x: x[-1]
res('bananas')
```

```python
nums = [1,2,3,4,5]
res = list(map(lambda x:x**2, nums))
res
```

```python
fruits = ["apple","banana","kiwi"]
res = list(map(lambda x: x.upper(), fruits))
res
```

```python
temps = [0,10,20,30]
res = list(map(lambda x: x*1.8+32, temps))
res
```

```python
random_words = ['yea', 'delicious', 'happy', 'insane']
res = list(map(lambda x: len(x), random_words))
res
```

```python
nums = [1,2,3,4,5,6,7,8,9]
res = list(filter(lambda x:x%2==0, nums))
res
```

```python
nums = list(range(40,55))
res = list(filter(lambda x: x>=50, nums))
res
```

```python
words = ['idkidfjdj', 'dfmdkf', 'dkf', 'f']
res = list(filter(lambda x:len(x)>=5, words))
res
```

```python
names = ['akshit', 'bob', 'charlie', 'akon']
res = list(filter(lambda x:x.startswith('a'), names))
res
```

```python
words = ["pear","apple","kiwi","banana"]
res = sorted(words, key=lambda x:len(x))
res
```

```python
nums = [-5,7,-2,10,-9]
res = sorted(nums, key=lambda x:abs(x))
res
```

```python
students = [
    ("Alice",82),
    ("Bob",95),
    ("Charlie",71)
]
res = sorted(students, key=lambda x:x[1])
res
```

```python
res = sorted(students, key=lambda x:x[0])
res
```

```python
words = ['nfcnfkvn', 'mfrigjtkfmkrmf', 'frifjrfr', 'dkjmkdjkfg']
res = max(words, key= lambda x:len(x))
res
```

```python
max(students, key=lambda x:x[1])
```

```python
nums = [-12, 4, -2, 9, 1, -7]
min(nums, key=lambda x:abs(x))
```

```python
employees = [
    {"name":"Alice","salary":5000},
    {"name":"Bob","salary":9000},
    {"name":"Charlie","salary":6500}
]
res = sorted(employees, key=lambda x:x['salary'], reverse=True)
res
```

```python
res = list(filter(lambda x:x['salary']>=6000, employees))
res
```

```python
res = list(map(lambda x:x['name'], employees))
res
```

```python
res = list(map(lambda x:x['salary']*12, employees))
res
```

```python
res = sorted(employees, key=lambda x:len(x['name']))
res
```

```python
fruits = ["apple","Pear","banana","Kiwi"]
res = sorted(fruits, key= lambda x:x.lower())
res
```

```python
idk = [(3,7),(1,5),(2,1),(4,8)]
res = sorted(idk, key=lambda x:x[1])
res
```

```python
words = ["apple", "education", "rhythm", "banana", "queue"]
max(words, key=lambda x:sum(1 for ch in x if ch in 'aeiou'))
```

```python
res = sorted(words, key=lambda x:x[-1])
res
```

```python
nums = [1,2,3,4,5,6]
res = list(filter(lambda x:x if x%2==0 else 0,(map(lambda x:x**2,nums))))
res
```

```python
#funcs can also be reversed
res = list(map(lambda x:x**2,(filter(lambda x:x if x%2==0 else 0,nums))))
res
```

## Reduce Function
### reduces iterable into one value

```python
from functools import reduce
```

```python
nums = [1,2,3,4]
val = reduce(lambda x,y:x+y, nums)
val
```

```python
nums = [2,3,4]
val = reduce(lambda x,y:x*y, nums)
val
```

```python
words=["Data","Science","AI"]
val = reduce(lambda x,y: x+y, words)
val
```

```python
nums=[8,5,12,7]
val = reduce(lambda x,y:x if x>y else y, nums)
val
```

```python
nums=[8,5,12,7]
val = reduce(lambda x,y: x if x<y else y, nums)
val
```

```python
words=["apple","banana","pear"]
val = reduce(lambda x,y:x+"-"+y, words)
val
```

```python
employees = [[101,'amit', 45000],
[102,'shalini', 35000],
[103, 'rohit', 25000],
[104, 'idk', 60000],
[105, 'idk2', 10]
]

user_in = int(input('Sort based on: 0. Employee No, 1. Name, 2. Salary: '))
uM  = sorted(employees, key=lambda x:x[user_in])
uM
```
