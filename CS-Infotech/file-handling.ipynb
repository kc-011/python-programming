# File Handling Practice

## Q. Print the longest word from a file

```python
with open('text.txt', 'r') as fp:
    lines = fp.readline()
    line = lines.split(' ')
    line_new = [x.strip(" ,.") for x in line]
max(line_new, key=len)
#OR you can sort and then use index to fetch it
```

## Q. Print only code and no comments

```python
with open('code.txt', 'r') as fp:
    lines = fp.readlines()
    new_lines = [line.rstrip("\n") for line in lines if not line.lstrip().startswith('#')]
    for line in new_lines:
        print(line)
```

## Q. Return no. of lines, words and characters in a file

```python
with open('text.txt', 'r') as fp:
    lines = fp.readline()
    no_of_chars = [line for line in lines]
    no_of_lines = lines.split('. ')
    no_of_words = lines.split()

print("Number of words:", len(no_of_words))
print("Number of lines:", len(no_of_lines))
print("Number of characters:", len(no_of_chars))
```

## Q. Calculate total balance from deposits & withdrawal

```python
with open('balance-sheet.txt', 'r') as fp:
    lines = fp.readlines()
    total = 0
    for line in lines:
        action, money = line.split(', ')
        if action == "D":
            total += int(money)
        else:
            total -= int(money)
print("Total balance:", total)
```
