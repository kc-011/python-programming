## Q. Print the longest word from a file

```python
with open('text.txt', 'r') as fp:
    lines = fp.readline()
    line = lines.split(' ')
    line_new = [x.strip(" ,.") for x in line]
max(line_new, key=len)
#OR you can sort and then use index to fetch it
```

## Q. Print only code and no comments

```python
with open('code.txt', 'r') as fp:
    lines = fp.readlines()
    new_lines = [line.rstrip("\n") for line in lines if not line.lstrip().startswith('#')]
    for line in new_lines:
        print(line)
```

## Q. Return no. of lines, words and characters in a file

```python
with open('text.txt', 'r') as fp:
    lines = fp.readline()
    no_of_chars = [line for line in lines]
    no_of_lines = lines.split('. ')
    no_of_words = lines.split()

print("Number of words:", len(no_of_words))
print("Number of lines:", len(no_of_lines))
print("Number of characters:", len(no_of_chars))
```

## Q. Calculate total balance from deposits & withdrawal

```python
with open('balance-sheet.txt', 'r') as fp:
    lines = fp.readlines()
    total = 0
    for line in lines:
        action, money = line.split(', ')
        if action == "D":
            total += int(money)
        else:
            total -= int(money)
print("Total balance:", total)
```

## Q. Make a marks analyzer
Calculate Total, Average and Grade & store it at the end of the line

```python
with open('marks.txt', 'r') as fp:
    lines = fp.readlines()
    idk = [line.strip().split(',') for line in lines]
    new_idk = [[word.strip() for word in inner] for inner in idk]
    subjects = new_idk[0]
    subjects.extend(['Total', 'Average', 'Grade'])
    students = new_idk[1:]
    
    total = 0
    new_lst = [subjects]
    for i in students:
        for j in range(1,len(i)):
            i[j] = int(i[j])
            total  += i[j]
        i.append(total)
        avg = total/(len(i)-2)
        i.append(avg)
        if avg>=70:
            i.append('A')
        elif avg>=60:
            i.append('B')
        elif avg>=50:
            i.append('C')
        elif avg>=40:
            i.append('D')
        else:
            i.append('F')
        total = 0
        new_lst += [i]
# just have to add the new stuff now

final_lst = [[str(i) for i in inner]for inner in new_lst]
with open('new_marks.txt', 'w') as fp:
    fp.seek(0)
    for i in final_lst:
        fp.write(", ".join(i)+"\n")
```
