### Tuples <br>
Immutable, heterogeneous and ordered

_Tuple Creation_

```python
#Defining an empty tuple
tup1 = ()
print(tup1)

#Using String
tup2 = ("Hey","Man")
print(tup2)

#Using List
list1 = [1,2,3,4,5]
print(tuple(list1))

#Using Built-in Function (can only pass 1 argument)
tup3 = tuple('Hey')
print(tup3)

#To define a tuple with a single element, you must put ',' after the element
tup4 = ("Sup",)
print(tup4)
```

_Tuple Operations_

```python
tup5 = (5, "Ok", True, 7)
print(tup5)
tup6 = ("Hey", "Hru",333)

#To make a nested tuple:
tup7 = (tup5,tup6)
print(tup7)

#Valid operations: + and *
tup8 = tup5 + tup6
print(tup8)

tup5 = tup5 + ("hi",)
print(tup5)

print(("Yoo",)*3)
```

_Tuple Indexing and Slicing_ - same as Lists<br>
<br>
The only difference is that you cannot do tup[index] == smth since they are immutable

```python
#We cannot delete an element from the tuple due to it being immutable so
# we can just delete the entire tuple instead

tup9 = (1,2,1,1,2,1)
del tup9

#print(tup9) - NameError: name 'tup9' is not defined
```

_Tuple Unpacking_

```python
#You can use normal loops to get values:
for i in tup2:
    print(i)

#or, you can unpack to get the values separately
tup10 = (10,20,30)
a,b,c = tup10
print(a)
print(b)
print(c)

#or, you can unpack with asterisk (*) to get multiple items in a list
tup11 = ["Apple", "Banana", "Orange", "Lemon", "Ginger"]
a, *b, c = tup11
print(a)
print(b) 
print(c)
#print(d) if we add one more, the * will balance out the number of items
```
