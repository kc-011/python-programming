*List Creation*

```python
#1
list3 = list("hey")
print(list3)
#2
list4 = []
#3
list5 = list3 + list4
#Or
list5 = list3 + [1] #or whatever you want
print(list5)
#4
list6 = list3 * 3
print(list6)

list1 = []
for i in range(10):
    list1 += [i-5]
print(list1)
```

*List Indexing and Slicing*

```python
print(list3[0])
print(list3[-2])
#list3[1][0] for nested lists

print(list3[2:])
print(list6[:7])
print(list6[4::-1])
print(list6[-6:-2])
print(list6[-1:-4])
print(list6[-1:-4:-1])
```

*List Methods* <br>
These modify the list and return 'None

```python
list7 = [1,"a", 944, True, "Idk lol"]
print(list7.append(5)) #will return None, so don't print methods straight away
print(list7)

list7.extend([4,5]) #only iterables
print(list7)

list7.insert(4, False)
print(list7)

list7.remove(5)
print(list7)

#list7.sort() only works if list has same type of variables in it
list3.sort(reverse=True) #default of reverse is False
print(list3)

list7.reverse()
print(list7)
```

*Mutation vs Reassignment*

```python
#Mutation
list8 = list7 
list8.append(9)
#This changes original object
print(list7)
print(list8)
#why? because both variables point to the SAME object
```

```python
#Reassignment
list9 = list8
list9 = list9 + ["lol"] 
#This creates a new object (og unchanged)
print(list8)
print(list9)
```

*The * Operator Rule*

```python
print([5]*3) 
print([[5]]*3) #this creates 3 references to the same object
# like A * 3 = [A,A,A]
# changing one will change all
list10 = [[5]] 
#^ [A]
list11 = list10*3 #it creates a new outer list, not inner ones (same references)
#^ [A, A, A]
list11[2][0] = 99
print(list10) #this also changes since it is one of the references in list11
print(list11) #contains 3 references to that SAME inner list
```

*List Comprehension*

[ _expression_  **for variable**  _in iterable_  **if condition** ]


```python
print([i for i in range(5)])
print([i**2 for i in range(6)])

print([i for i in range(10) if i % 3 == 0])
print([i/2 for i in range (10) if i % 3 == 0])
```

The above is used for _filtering_ <br>
If we want to replace values instead, the expression would look like:<br>
[**if** _something_ **else** _something_ **for variable** _in iterable_] <br>
Example -> [i if i % 2 == 0 else -1 for i in range(5)]

```python
print([i if i > 2 else -1 for i in range(8)])
print([i*3 if i %2 == 0 else i for i in range(10)])
```

_Nested List Comprehension_

```python
print([(i, j) for i in range(2) for j in range(3)])
print([(i**2,j**2) for i in range(3,6) for j in range(2,5)])
```

```python
matrix = [[0 for i in range(3)] for i in range(3)]
print(matrix)
#this avoids the [[0]*3]*3 bug 
```
