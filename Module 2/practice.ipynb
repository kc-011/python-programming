## Q. Input 10 random numbers in a list
### Without repitition <br>

```python
import random
lst = []
while len(lst)!=10:
    num = random.randint(1,20)
    if num not in lst:
        lst.append(num)
print(lst)
```

## Q. Find the mean and median of a list of student's marks

```python
marks = list()
for i in range(1,11):
    marks.append(random.randint(1,100))
print(marks)
```

```python
mean_marks = sum(marks)/len(marks)
print(f"Mean Marks: {mean_marks}")
```

```python
marks.sort()
print(marks)
length_marks = len(marks)
if length_marks%2==0:
    median_marks = (marks[length_marks//2] + marks[(length_marks//2)+1])/2
else:
    median_marks = marks[(length_marks//2)+1]
print(f"Median Marks: {median_marks}")
```
