## Sets
### Immutable*, Unordered, No duplicates and Heterogeneous

*Items can be deleted/added after creation

Sets Creation

```python
set1 = set((1,3,5,7,8,9))
print(set1)

list1 = [1,2,3,3,4,5]
set2 = set(list1)
print(set2)

set3 = {"hi", "yo", 5}
print(set3)

set4 = {x**2 for x in range(1,11)}
print(set4)
```

Adding items in a Set

```python
set5 = {"apple", "banana", "cherry"}
set5.add("orange")
print(set5)

list2 = [True, 1, 0] 
set5.update(list2)
print(set5)
```

Removing items in a Set

```python
set6 = {1,2,3,4,5}
set6.remove(5)
print(set6)

#remove() will raise a KeyError if item does not exist in Set

set6.discard(3)
print(set6)

#discard() will not raise an error if item is not found
set6.discard(5)

#we can also use pop() but since sets are unordered, we don't know which item will get removed
x = set6.pop()
print(x)

#clear() empties the whole set
set6.clear()
print(set6)

#del will delete the set completely
```

### Set Methods

Since sets are similar to sets in maths, we can do multiple set theory operations on them like union, intersection, etc

Union 

```python
#Union - joins 2 (or more) sets and returns a new set 

print(set2.union(set3))
print(set3.union(set4, set5, set6))

#union() can take any iterable and join it
print(set3.union(list2))

#We can also use | (or) symbol BUT it will only work for sets and no other iterables
print(set2 | set3) #can put more like | set4 | set5 etc

#We can also use update() func but it updates the old set, does not return a new set
```

Intersection

```python
#Intersections keeps only the common stuff

print(set1.intersection(set2))

#same as union, intersection() can take any iterable
print(set1.intersection(set2, list1))
#and for SETS ONLY, we can use the & operator
print(set2 & set3 & set1)
#To update the original list, we can use:
set1.intersection_update(set2)
print(set1)
```

Difference

```python
#returns a set of items that are present in set A but not in set B (or like set A - set B)
set8 = {"apple", "banana", "cherry"}
set9 = {"google", "microsoft", "apple"}
set10 = {"microsoft", "banana"}
print(set8.difference(set9))
#the above func can be used with any no. of iterables
print(set8.difference(set9, set10))

#we can also use the '-' operator (with sets ONLY)
print(set8 - set9)

#difference.update() will edit the original list instead of making a new one
```

Symmetric Difference

```python
#Returns a set of items that are NOT present in both sets
# set8 = {"apple", "banana", "cherry"}
# set9 = {"google", "microsoft", "apple"}
# set10 = {"microsoft", "banana"}
# for reference^

print(set8.symmetric_difference(set9)) #can only take ONE argument
print(set8.symmetric_difference(set10)) 

#we can also use the ^ symbol (ONLY for sets)
print(set9 ^ set10)

#symmetric_difference_update() updates the original set
```

Frozen Sets

```python
#They are sets that cannot be changed at all - no items can be added or removed
set11 = frozenset({"Hi", "Hey", "Yo", 44, True})
print(set11)

#frozensets support all non-mutating functions of sets:
#copy()	 	
#difference() -	
#intersection()	&	
#isdisjoint()	 	Returns whether two frozensets have an intersection	
#issubset()	<= / <	Returns True if this frozenset is a (proper) subset of another	
#issuperset()	>= / >	Returns True if this frozenset is a (proper) superset of another	
##union()	|

#As you can see above, none of the update() functions are there because the original (frozen)set cannot be changed
```
