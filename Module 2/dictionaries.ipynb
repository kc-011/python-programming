## Dictionaries
#### Ordered and Mutable and Heterogeneous

Dictionary Creation

```python
dict1 = {}
print(type(dict1))
dict2 = dict()
print(type(dict2))
dict3 = dict(a = "Hi", b = "HiHi") #can only use a,b,c.. as keys in dict()
print(dict3)
dict4 = {"name": "Karan", "age":22}
print(type(dict4))
```

Accessing a Dictionary

```python
print(dict4["name"])
#or
print(dict4.get("age"))
```

Adding and Updating items 

```python
dict5 = {"Name": "Rahul", "age": 20, (1,2,3): [1,2,3]}

dict5["height"] = 5.7 #adding a key, value
print(dict5)

dict5["age"] = 21 #changing a value
print(dict5)

#Keys cannot be updated - they are considered immutable and unique, but can be deleted
```

Iterating through a Dictionary

```python
dict8 = {"maths" : 95, "biology" : 78, "physics" : 86, "chemistry" : 88, "english" : 90}

for key in dict8: #to print keys
    print(key)

for value in dict8.values(): #to print values
    print(value)

for key, value in dict8.items(): #to print keys & values
    print(f"{key}:{value}")
```

Nested Dictionary

```python
d = {1: 'Geeks', 2: 'For', 3: {'A': 'Welcome', 'B': 'To', 'C': 'Geeks'}}
print(d)

#If we want the value of B:
print(d[3]["B"])

#let's try to get the values of the nested list
for i in d[3].values():
   print(i)

#that was easier than expected...

for i in d[3]: #for keys in the nested dictionary
    print(i)
```

```python
#another way to access nested dictionaries
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}

for x, obj in myfamily.items():
    print(x) #print the outer keys
    
    for y in obj: #for each outer key's value, print the inner keys and its values
        print(y + ':', obj[y])
```
