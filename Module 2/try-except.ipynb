## Try-Except Blocks

```python
try:
    num = int(input("Enter a number: "))
    print(f"Number entered: {num}")
except ValueError:
    print("Please enter an integer")
```

```python
while True: #program keeps asking until user inputs valid stuff
    try:
        num1 = int(input("Enter number 1: "))
        num2 = int(input("Enter number 2: "))
        print(num1/num2)
        break
    except ValueError:
        print("Please enter an integer only")
    except ZeroDivisionError:
        print("Cannot divide by 0")
#It is better to check for specific errors instead of all at once as it may hide bugs
```

```python
#Used for debugging, so we can see what type of error may come
try:
    num3 = int(input("Enter an integer: "))
    print(100/num3)
except Exception as e:
    print(f"This error occured: {e}")
```

```python
age = int(input("Enter age: "))

if age < 0:
    raise ValueError("Age cannot be negative")

print("Age:", age)
```

```python
try:
    num4 = int(input("Enter your age: "))
except ValueError:
    print("Please enter an integer")
else:
    if num4<0:
        raise ValueError("Age cannot be negative")
    else:
        print(f"Age: {num4}")
```

```python
#Lets make a program where user inputs a number, and we have to check if its not negative, in the range and an integer.
# If it satisfies the above, it will be stored in the lucky draw ticket for that user, otherwise it will ask the
# user to enter the number again
while True:
    try:
        user_num = int(input("Enter any number between 1 and 100 (both inclusive): "))

    except ValueError:
        print("Please enter a number")
        continue

    if user_num < 1 or user_num > 100:
        print("Number is not between 1 and 100")
        continue
           
    print("Your lucky number has been stored. The winner will be revealed soon")
    break
#Now let's add more stuff, let's see if the user wins the lucky draw
import random
import time
time.sleep(2)
print("The lucky number is now being chosen...")
time.sleep(3)
lucky_num = random.randint(1,100)
print("The lucky number is...")
time.sleep(2)
print(lucky_num)
time.sleep(1)
if lucky_num == user_num:
    print("CONGRATULATIONS!! You just won the lucky draw!!")
else:
    print(f"Your number was {user_num}. Sorry (¬_¬ ) better luck next time")
```

In the above code, I did not use an except block with my condition (num being between 1 and 100) because thats MY requirement, I can just catch it with a simple 'if' block. But things that are out of my control like the user inputting a string or a float, must be caught by the program itself which is why there is only one except block. These things are not meant to be overused and should be separated from your own rules
