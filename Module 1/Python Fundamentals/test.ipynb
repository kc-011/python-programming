```python
x = int(input("Enter a number: ")) 
divisor = 1
count = 0
while divisor <= x: 
    if x % divisor == 0:
        count +=1 
        divisor+=1 
        if count > 2:
            print(f"{x} is not a prime number")
            break
        elif count < 2 and divisor == x:
            print(f"{x} is a prime number")
            #break
    else:
        divisor+=1
```

```python
n =  999983 

if n <= 1:
    print("Not Prime")
else:
    i = 2
    while i * i <= n:
        if n % i == 0:
            print("Not Prime")
            break
        i += 1
    else:
        print("Prime")
```

```python
def countdown(n):
  if n <= 0:
    print("Done!")
  else:
    print(n)
    countdown(n - 1)

countdown(5)
```

```python
def fibonacci(n):
  if n <= 1:
    return n
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(7))
```
