## Q. Compress a name
E.g - Karan Chawla --> K.Chawla <br>
Karan --> Karan <br>
Karan Singh Chawla --> K.S.Chawla <br>

Separate and put into a list<br>
Slice all names except last just to get first letter <br>
Then concatenate to new string <br>
Then at the end, add the last name <br>

```python
userinput1 = input('Enter a name: ')
name_list = userinput1.split(' ')
compressed1 = ''
for i in range(len(name_list)-1):
    compressed1=compressed1+name_list[i][:1].capitalize()+'.'
compressed1+=name_list[-1].title()
print(compressed1)
```

## Q. Decompress a String

a3b2c3d1 --> aaabbcccd

Logic: <br>
Iterate through the string using a for loop <br>
Loop at increments of 2, to read the letter <br>
Read the char that is next to the letter, which is its number <br>
Multiply the letter with that number and concat to an empty string <br>

```python
userinput2 = input('Enter a compressed string: ')
new_str = ''
for i in range(0, len(userinput2)-1, 2):
    char = userinput2[i]
    num = userinput2[i+1]
    new_str=new_str+char*int(num)
print(new_str)
```
