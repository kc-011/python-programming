## Z-Test

```python
import numpy as np
from scipy.stats import norm
from scipy import stats
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(n))
z_score
```

```python
p_value = 2*(1-norm.cdf(abs(z_score)))
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("Null hypothesis rejected")
else:
    print("Null hypothesis accepted")
```

## T-Test

```python
#sample and sample mean we alr have
#but we need sample std now not population std
sample_std = np.std(sample, ddof=1) #degree of freedom
```

```python
t_stats = (sample_mean - population_mean)/(sample_std/np.sqrt(n))
t_stats
```

```python
p_value_1 = 2*(1-stats.t.cdf(abs(t_stats), df=n-1))
alpha=0.05
p_value_1
```

```python
if p_value_1 < alpha:
    print("Null hypothesis rejected")
else:
    print("Null hypothesis accepted")
```

## Two Sample Test

```python
boys = [85,88,90,92,87,85,89,91,86,88]
girls = [82,84,80,83,81,79,78,85,84,83]
#Null Hypo - No difference
#Alternate Hypo - Significant difference b/w performance
```

```python
t_stats_1, p_value_2 = stats.ttest_ind(boys, girls, equal_var=False)
print(t_stats_1)
print(p_value_2)
```

```python
if p_value_2 < alpha:
    print("Null hypothesis rejected")
else:
    print("Null hypothesis accepted")
```

## Chi Test

```python
import seaborn as sns
from scipy.stats import chi2_contingency
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df
```

```python
contigency_table = pd.crosstab(df['sex'], df['survived'])
contigency_table
```

```python
chi2, p_value_3 , dof, expected = chi2_contingency(contigency_table)
print(chi2)
print(p_value_3)
print(dof)
print(expected)
```

```python
if p_value_3 < alpha:
    print("Null hypothesis rejected - There is a significant difference")
else:
    print("Null hypothesis accepted - There is no difference")
```

## Annova Test

Analysis of Variance - used to compare means of more than two independent groups to check if at least one group mean is significantly different

E.g - "Do different teaching methods result in different average scores?" 

```python
from scipy.stats import f_oneway
```

```python
df1 = sns.load_dataset('titanic')
df1
#we want to check if theres any relation b/w age and passenger class
```

```python
df1 = df1[['pclass', 'age']].dropna()
```

```python
df1['pclass'].unique()
```

```python
class_1 = df1[df1['pclass']==1]['age']
class_2 = df1[df1['pclass']==2]['age']
class_3 = df1[df1['pclass']==3]['age']
class_1
```

```python
f_stats , p_value_4 = f_oneway(class_1, class_2, class_3)
print(f_stats)
print(p_value_4)
```

```python
if p_value_4 < alpha:
    print("Null hypothesis rejected - There is a significant difference")
else:
    print("Null hypothesis accepted - There is no difference")
```
