## Z-Test

```python
import numpy as np
from scipy.stats import norm
from scipy import stats
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(n))
z_score
```

```python
p_value = 2*(1-norm.cdf(abs(z_score)))
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("Null hypothesis rejected")
else:
    print("Null hypothesis accepted")
```

## T-Test

```python
#sample and sample mean we alr have
#but we need sample std now not population std
sample_std = np.std(sample, ddof=1) #degree of freedom
```

```python
t_stats = (sample_mean - population_mean)/(sample_std/np.sqrt(n))
t_stats
```

```python
p_value_1 = 2*(1-stats.t.cdf(abs(t_stats), df=n-1))
alpha=0.05
p_value_1
```

```python
if p_value_1 < alpha:
    print("Null hypothesis rejected")
else:
    print("Null hypothesis accepted")
```
