# Matplot Library

```python
import matplotlib.pyplot as plt
#print(matplotlib.__version__)
```

## Line Plots

```python
import numpy as np
x_p = np.array([1,8]) # x-axis
y_p = np.array([3,10]) # y-axis

plt.plot(x_p, y_p)
plt.show()
```

```python
plt.plot(x_p, y_p, 'o')
plt.show()
```

```python
x1_p = np.array([1,3,6,8])
y1_p = np.array([3,8,1,10])
#the x1 values correspond w/y1 values --> (1,3), (3,8), etc
plt.plot(x1_p, y1_p)
plt.show()
```

```python
y2_p = np.array([3,8,1,10,5,7])
plt.plot(y2_p)
plt.show() #graph gives default x values (0,1,2,3...) depending on length of y values
```

## Markers

```python
plt.plot(y2_p, marker = 'd') #can put marker to make points more visible
plt.show() #there are alot of markers (read abt it)
```

### Marker Size & Color

```python
plt.plot(y2_p, marker = 'h', ms= 20, mec = 'k', mfc = 'm')
#mec = marker edge color
#mfc = marker face color
#can use hex codes also
plt.show()
```

We can change the style of the line using <i>linestyle</i> or <i>ls</i><br>
And we can change it to: <br>
~ dotted or : <br>
~ dashed or -- <br>
~ solid or - (default) <br>
~ dashdot or -. <br>
~ None or ""

We can change the color of the line too using <i>color</i> or <i>c</i> with: <br>
~ 140 supported color names
~ Hex code

We can also change the width of the line using <i>linewidth</i> or <i>lw</i>

### Multiple Lines

```python
y3_p = np.array([3,8,1,10])
y4_p = np.array([6,2,7,11])
plt.plot(y3_p, linestyle = 'dashed', color = 'OliveDrab', linewidth = 4)
plt.plot(y4_p, ls = ':', c = '#4034eb', lw = 6)

plt.show()

#we can also put multiple arrays in a single plot function
```

### Format strings (fmt)

It's a shortcut string notation to specify the below things: <br> 
Syntax: <i>marker|line|color<i>

```python
y3_p = np.array([3,8,1,10,4])
plt.plot(y3_p, 's:g') #square(s)|dotted lines(:)|green(g)
plt.show()
#There are a few line designs such as:
# '-' --> Solid line
# ':' --> Dotted line
# '--' --> Dashed line
# '-.' --> Dashed/dotted line

#If you leave out the line parameter in fmt, no line will be shown
#You can choose the color by writing the first letter of the color
```

## Graph Labels

```python
x2_p = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y2_p = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x2_p, y2_p)

font1 = {'family':'Arial', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color': 'darkred', 'size':15}

plt.title("Apple Watch Data", fontdict=font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.show()
```

## Displaying Multiple Plots (Subplots)

```python
# in the subplot() func, we have to provide 3 arguments that specify the row, column and index (not starting from 0)

#if we want the graphs next to each other/side by side:
plt.subplot(1,2,1)
plt.plot(y3_p)

plt.subplot(1,2,2)
plt.suptitle("IDK") #super title
plt.plot(y4_p)

plt.show()
```

```python
# if we want the graphs to go on top of each other:
plt.subplot(2,1,1)
plt.plot(y3_p)

plt.subplot(2,1,2)
plt.plot(y4_p)

plt.show()
```

We can draw as many plots as we want on one figure, just have to specify row, column, and index. for example, if we want 6 figures in one frame, with 2 rows and 3 columns, we'll do - (2,3,1), (2,3,2), (2,3,3), (2,3,4)... etc.

# Scatter Plots

```python
x3 = np.array([12,4,20,17,14,3,8,1,9,18,15,10,17,16,4])
x4 = np.array([99,88,68,97,67,88,67,68,87,57,84,67,46,75,68])

plt.scatter(x3,x4)
plt.show()
```

```python
#We can merge 2 scatter plots to see the relation between stuff
x3 = np.array([12,4,20,17,14,3,8,1,9,18,15,10,17,16,4])
x4 = np.array([99,88,68,97,67,88,67,68,87,57,84,67,46,75,68])
plt.scatter(x3,x4)

x5 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y5 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x5,y5)

plt.show()
#the default colors are blue & orange (idk why)
```

We can also set the color of each scatter plot using the <i>color</i> or <i>c</i> argument

```python
plt.scatter(x3,x4, color = 'red')
plt.scatter(x5,y5, c = '#e62ce6')
plt.show()
```

We can also define the color of each dot, but only using <i>c</i>

```python
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta", "red", "green"])
plt.scatter(x5,y5, c = colors)
plt.show()
```

```python
#colors2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x5,y5, c=y5 ,cmap='twilight')
plt.colorbar()
plt.show()
```

We can also change the size of the dots using <i>s</i>
Just like color, the array also must be the same size as the x and y axis OR just give 1 size to all

The transparency can be changed too using the <i>alpha</i> argument. The array must be the same length as x & y axis arrays OR just give one alpha number to all

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75,300,80])
plt.scatter(x5,y5,s=sizes, alpha=0.5)
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

## Bar/Histogram/Pie Plots

<b>Bar Graph</b>

~<i>plt.bar</i> for vertical bars<br>
~<i>plt.barh</i> for horizontal bars <br>
~<i>color</i> argument stays the same <br>
~To change width: use <i>width</i> argument <br>
~If it's <i>plt.barh</i> --> then use <i>height</i> argument 

```python
x6 = np.array([4,5,6,8,9,3,2])
y6 = np.array([20,56,37,86,53,90,66])
colors = ["blue", "green", "yellow", "orange", "red", "pink", "purple"]

plt.subplot(2,1,1)
plt.bar(x6,y6, color=colors, width=0.8)
plt.show()

plt.subplot(2,1,2)
plt.barh(x6,y6, color='DarkTurquoise', height=0.4)
plt.show()
```

<b>Histograms</b> <br>

~Shows frequency distributions <br>
~Meaning that it gives number of observations within each given (or calculated) interval. For example: <br>
2 people from 140 to 145 cm <br>
7 people from 145 to 150 cm <br>
11 people from 150 to 155 cm <br>
etc....<br>
~<i>plt.hist</i> is used

```python
random_x = np.random.normal(170,10,250) 
#this means 250 values that will be around 170 and have standard deviation of 10
plt.hist(random_x, color='LightCoral')
plt.show()
```

<b>Pie Chart</b>

~Command: <i>plt.pie</i><br>
~A piece of the pie is called a wedge. The first wedge starts from the x-axis and anti-clockwise <br>
~We add labels to each wedge using the <i>labels</i> parameter by passing an array<br>
~We can also change the start angle of the wedge. Default is 0 degrees on x-axis but can be changed to any degree <br>
~If you want a wedge to stand out, you can do so with the <i>explode</i> parameter, where you have to pass an array of values for each wedge. For e.g - [0.2, 0, 0, 0] (if there are 4 wedges, respectively)<br>
~Shadows can be added to the pie by setting the <i>shadows</i> parameter to <i>True</i><br>
~Colors paramter is the same as before<br>
~We can use the <i>plt.legend()</i> function to show the legend and a title for it using the <i>title</i> paramter inside the legend function<br>




```python
pie_x = np.array([35,45,60,80])
label = ["Apples", "Oranges", "Grapes", "Bananas"]
explode_x = np.array([0.2,0,0,0])
colors3 = ["Green", "Purple", "Orange", "red"]
plt.pie(pie_x, labels=label, colors=colors3, startangle=90,explode=explode_x, shadow=True)
plt.legend(title="Fruits")
plt.show()
```
