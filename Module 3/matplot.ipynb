# Matplot Library

```python
import matplotlib.pyplot as plt
#print(matplotlib.__version__)
```

```python
import numpy as np
x_p = np.array([1,8]) # x-axis
y_p = np.array([3,10]) # y-axis

plt.plot(x_p, y_p)
plt.show()
```

```python
plt.plot(x_p, y_p, 'o')
plt.show()
```

```python
x1_p = np.array([1,3,6,8])
y1_p = np.array([3,8,1,10])
#the x1 values correspond w/y1 values --> (1,3), (3,8), etc
plt.plot(x1_p, y1_p)
plt.show()
```

```python
y2_p = np.array([3,8,1,10,5,7])
plt.plot(y2_p)
plt.show() #graph gives default x values (0,1,2,3...) depending on length of y values
```

## Markers

```python
plt.plot(y2_p, marker = 'd') #can put marker to make points more visible
plt.show() #there are alot of markers (read abt it)
```

### Marker Size & Color

```python
plt.plot(y2_p, marker = 'h', ms= 20, mec = 'k', mfc = 'm')
#mec = marker edge color
#mfc = marker face color
#can use hex codes also
plt.show()
```

## Line Plots

We can change the style of the line using <i>linestyle</i> or <i>ls</i><br>
And we can change it to: <br>
~ dotted or : <br>
~ dashed or -- <br>
~ solid or - (default) <br>
~ dashdot or -. <br>
~ None or ""

We can change the color of the line too using <i>color</i> or <i>c</i> with: <br>
~ 140 supported color names
~ Hex code

We can also change the width of the line using <i>linewidth</i> or <i>lw</i>

### Multiple Lines

```python
y3_p = np.array([3,8,1,10])
y4_p = np.array([6,2,7,11])
plt.plot(y3_p, linestyle = 'dashed', color = 'OliveDrab', linewidth = 4)
plt.plot(y4_p, ls = ':', c = '#4034eb', lw = 6)

plt.show()

#we can also put multiple arrays in a single plot function
```

### Format strings (fmt)

It's a shortcut string notation to specify the below things: <br> 
Syntax: <i>marker|line|color<i>

```python
y3_p = np.array([3,8,1,10,4])
plt.plot(y3_p, 's:g') #square(s)|dotted lines(:)|green(g)
plt.show()
#There are a few line designs such as:
# '-' --> Solid line
# ':' --> Dotted line
# '--' --> Dashed line
# '-.' --> Dashed/dotted line

#If you leave out the line parameter in fmt, no line will be shown
#You can choose the color by writing the first letter of the color
```

## Graph Labels

```python
x2_p = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y2_p = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x2_p, y2_p)

font1 = {'family':'Arial', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color': 'darkred', 'size':15}

plt.title("Apple Watch Data", fontdict=font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.show()
```

### Grid Lines

```python
plt.plot(y3_p)
#can choose which axis (default is both), and the below)
plt.grid(axis='y', color='navy', linestyle='--', linewidth=2)
plt.show()
```

## Displaying Multiple Plots (Subplots)

```python
# in the subplot() func, we have to provide 3 arguments that specify the row, column and index (not starting from 0)

#if we want the graphs next to each other/side by side:
plt.subplot(1,2,1)
plt.plot(y3_p)

plt.subplot(1,2,2)
plt.suptitle("IDK") #super title
plt.plot(y4_p)

plt.show()
```

```python
# if we want the graphs to go on top of each other:
plt.subplot(2,1,1)
plt.plot(y3_p)

plt.subplot(2,1,2)
plt.plot(y4_p)

plt.show()
```

We can draw as many plots as we want on one figure, just have to specify row, column, and index. for example, if we want 6 figures in one frame, with 2 rows and 3 columns, we'll do - (2,3,1), (2,3,2), (2,3,3), (2,3,4)... etc.
