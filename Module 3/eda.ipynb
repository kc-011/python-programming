### First, we'll look at some columns individually (Univariate)

Age groups of all Genders

```python
sns.set_context('paper')
sns.countplot(x=df['Age'], hue=df['Gender'])
```

Screen Time (hrs/day)

```python
sns.histplot(data=df, x='Screen Time (hrs/day)', alpha=0.8)
sns.set_context('talk')
```

Sleep Duration (hrs)

```python
sns.boxplot(x=df['Sleep Duration (hrs)'], hue=df['Gender'], legend='full')
sns.set_context('paper')
```

Physical Activity (hrs/week)

```python
sns.boxplot(x=df['Physical Activity (hrs/week)'], hue=df['Gender'])
```

Percentage of Stress Levels

```python
sns.countplot(data=df, x='Stress Level', stat='percent')
```

Count of people feeling anxious before exams

```python
sns.countplot(x=df['Anxious Before Exams'], stat='proportion')
```

## Observations

1. The age ranges from 15 to 26 years old <br>
2. The average sleep hours (for all genders) is between 6-7 hours <br>
3. The average number of physical activity per week (~5 hrs) is same for all genders<br>
4. Nearly 50% have a medium stress level, followed with low stress levels (~33%) and then high stress levels with the lowest percentage (~18%) <br>
5. Almost half of the people don't get anxious before exams, whereas about 52% of them do

Let's group the above data into School, UG and PG and then compare

```python
df["Education Category"] = "Other"

df.loc[df["Education Level"].str.contains("Class", case=False), "Education Category"] = "School"
df.loc[df["Education Level"].str.startswith("B"), "Education Category"] = "Undergraduate"
df.loc[df["Education Level"].str.startswith("M"), "Education Category"] = "Postgraduate"
```

```python
df["Education Category"].value_counts()
```

```python
sns.countplot(data=df, x="Stress Level", hue="Education Category")
```

Postgrad students have the highest count in all levels of stress
