### First, we will look at our single variable plots

Age groups of all Genders

```python
sns.set_context('paper')
sns.countplot(x=df['Age'], hue=df['Gender'])
```

Screen Time (hrs/day)

```python
sns.histplot(data=df, x='Screen Time (hrs/day)', alpha=0.8)
sns.set_context('talk')
```

Sleep Duration (hrs)

```python
sns.boxplot(x=df['Sleep Duration (hrs)'], hue=df['Gender'], legend='full')
sns.set_context('paper')
```

Physical Activity (hrs/week)

```python
sns.kdeplot(x=df['Physical Activity (hrs/week)'], color='Indigo', fill=True, hue=df['Gender'])
```

Percentage of Stress Levels

```python
sns.countplot(data=df, x='Stress Level', stat='percent')
```
