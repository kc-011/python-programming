```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
penguins["island"].value_counts()
```

```python
sns.set_style("whitegrid") #default is tickers
#sns.set_context("talk") #default is notebook
sns.scatterplot(data=penguins, x="flipper_length_mm", y = "body_mass_g", hue = "island", palette="Set2")
#sns.despine() #removes the spines (idk)
```

```python
sns.scatterplot(data=penguins, x="species", y="body_mass_g", hue="island", style="sex", alpha=0.4)
```

```python
sns.stripplot(data=penguins, x="species", y="body_mass_g", hue="island", dodge=True)
```

```python
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island")
```

```python
sns.histplot(data=penguins, x="body_mass_g", hue="sex", alpha=0.8, multiple="stack")
```

```python
#regression plot
sns.set_style("ticks")
sns.regplot(data=penguins, x="body_mass_g", y="flipper_length_mm", color="black")
```

```python
# line plot
sns.set_style("whitegrid")
sns.lineplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="island", style="sex")
```

```python
#join plots
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm",kind="reg")
```

```python
# bar plot
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["pink", "skyblue"], estimator=np.var)
```

```python
#count plot
sns.countplot(data=penguins, x="species", hue="island", palette="Set2")
```

```python
#box plot
sns.boxplot(data=penguins, x="species",y="body_mass_g", hue="sex")
```

```python
#violin plot
sns.set_context("notebook")
sns.set_style("whitegrid")
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex", split=True, inner="box", alpha=1)
```

```python
#we can also put a swarm plot inside a violin plot
sns.violinplot(data=penguins, x="species", y="body_mass_g", palette="pastel")
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island",size=3, palette="dark")
```

```python
#kde plot (kernel density estimation) --> smooth version of a histogram
sns.kdeplot(data=penguins, x="body_mass_g", hue="species", fill=True)
```

```python
#Heatmap: only numeric columns
columns = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
penguins[columns].corr()
```

```python
sns.heatmap(data=penguins[columns].corr(), annot=True, cmap="Blues", vmin=-0.2, linewidths=2,linecolor="black")
```

```python
#Rug plot:
sns.rugplot(data=penguins, x="body_mass_g", hue="species", palette="Set2", height=2)
```

```python
#Pair plot
sns.pairplot(data=penguins, hue="sex", palette="Set2", diag_kind="hist")
```

```python
# Pair Grid
graph = sns.PairGrid(data=penguins, hue="sex", palette="Set2")
graph.map_upper(sns.scatterplot) #upper triangle
graph.map_lower(sns.kdeplot) #lower triangle
graph.map_diag(sns.histplot) #diagonal
graph.add_legend()
```
