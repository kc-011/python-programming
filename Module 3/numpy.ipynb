## NumPy

### Creating Arrays

```python
import numpy as np

#0D array
arr0 = np.array(45)

print(arr0)
print(type(arr0))

#1D array
arr1 = np.array([1,2,3,4,5])
print(arr1)

#2D array
arr2 = np.array([[1,3,5],[2,4,6]])
print(arr2)

#3D array
arr3 = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print(arr3)

#When creating an array, we can define the number of dimensions it has
arr4 = np.array([1,2,3], ndmin=5)
print(arr4)
```

### Array Creation using Functions

```python
arr5 = np.zeros((2,1)) #tuple of rows, columns - known as shape
print(arr5) #we can use to initialize an empty matrix b4 filling it with values

arr6 = np.ones((2,2)) #same as zeros
print(arr6) #often used in ML as test matrices

arr7 = np.empty((2,3)) #same as above
print(arr7) #contains random leftover (uninitialized) values from memory. faster than above 2 funcs

arr8 = np.arange(2,10,2) #same as normal range func
print(arr8)

arr9 = np.linspace(0,10,5) #creates an array of n(5) numbers between 0 and 10 (inclusive)
print(arr9)

arr10 = np.eye(3) #creates an identity matrix of n number of rows
print(arr10)

arr11 = np.random.rand(2) #(shape) 
print(arr11) #creates random values of given shape

arr12 = np.random.randint(2,12,(2,2))
print(arr12) #creates random values between given range of given shape
```

### Accessing Array Elements

It is the same as lists, but instead of doing list[1][2] for nested lists, we do arr[1, 2] in numpy

### Array Slicing 

It is also the same as list slicing but a bit different when it comes to nested arrays

```python
#1st number is to access the nested array(s), 2nd is to slice
print(arr2[1, 1:4]) #nested array [1] slice 1:4
print(arr2[0:2, 0:4]) #slice 0:4 in all nested arrays of index 0 and 1 
print(arr2[0:2, 2]) #return index 2 element from both nested arrays
```

### Shape Manipulation

```python
arr16 = np.array([[1,2],[3,4]])
print(arr16.reshape(2,1,2))
print(arr16.flatten()) #make it 1D
#or
print(arr16.reshape(-1))
print(arr16.T) #Transpose (swap rows and columns)
print(arr16.shape) #returns dimensions (rows, columns)
print(arr16.ndim) #returns just normal dimension
```

### Aggregate Operations

```python
arr13 = [[1,2,3],[4,5,6]]
print(np.sum(arr13))
print(np.mean(arr13))
print(np.min(arr13))
print(np.max(arr13))
print(np.var(arr13)) #variance - how much data spreads from the mean
print(np.std(arr13)) #root of variance

print(np.sum(arr13, axis=0)) #column-wise
print(np.sum(arr13, axis=1)) #row-wise
```

### Element-wise Functions

```python
print(np.sqrt(arr13))
print(np.exp(arr13)) #e raised to the power of each element
print(np.log(arr13)) #ln (element) aka log base e
print(np.abs(arr13)) #absolute value
```
