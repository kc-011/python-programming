```python
import pandas as pd
import yfinance as yf
idk = yf.download(tickers="AAPL", interval="15m", start="2026-03-31", end="2026-04-01")

```

```python
aapl_values = []
for i in idk.iloc[:, 0:1]:
    for j in idk[i]:
        aapl_values.append(j)
print(aapl_values)
#min_val = min(aapl_values)
#max_val = max(aapl_values)
```

```python
scaled_aapl = []
for i in aapl_values:
    min_val = min(aapl_values)
    max_val = max(aapl_values)
    scale = i - min_val
    range_ = max_val - min_val
    pos_ = scale/range_
    map_ = pos_ * 19
    new_y = round(19 - map_)
    scaled_aapl.append(new_y)
print(scaled_aapl)
len(scaled_aapl)
```
