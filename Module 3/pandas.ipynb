## Pandas

### Series <br>
1D Labeled array, heterogeneous

### Creating Series using List

```python
import pandas as pd
print(pd.__version__)

data1 = [100, 102, 104]
data2 = [103.5, 104.4, 110.9]
data3 = ["abc", True, 99, None]

series1 = pd.Series(data1)
#we can also give custom labeling (it must match no. of elements in data)
series2 = pd.Series(data2, index=["a", "b", "c"])
series3 = pd.Series(data3)

print(series1)
print(series2)
print(series3)
```

### Creating Series using Dictionary

```python
dict1 = {"Day 1": 1800, "Day 2": 1950, "Day 3": 1600, "Day 4": 2100}

series4 = pd.Series(dict1)
print(series4)
```

### Accessing Values <br>
loc, iloc, filtering, Boolean indexing

```python
#we normally do:
print(series1[1])
#but this is ambiguous - meaning we dont know if [1] is referring to index or label
#which is why we use loc and iloc
print(series2.loc["a"]) #based on label
print(series1.iloc[2]) #based on index

#we can also change a value using loc/iloc

series3.loc[3] = 5
series3.iloc[1] = False

print(series3)

#series4["Day 3"] += 400 
#print(series4)

#we can also do filtering/ boolean indexing like in numpy
print(series4[series4>=2000]) 
print(series4 < 2000)
```

```python
data4 = {"Name": ["Spongebob", "Patrick", "Squidward"], "Age": [30,35,50]}

df1 = pd.DataFrame(data4, index=["Employee 1", "Employee 2", "Employee 3"])
print(df1)

print(df1.loc["Employee 1"])
print(df1.iloc[2])

#Add a new column
df1["Job"] = ["Cook", "Intern", "Cashier"]
print(df1)

#Add a new row
new_row = pd.DataFrame([{"Name": "Sandy", "Age":28, "Job": "Engineer"}], 
                       index=["Employee 4"])
df1 = pd.concat([df1, new_row])
print(df1)

#Add new rows
new_rows = pd.DataFrame([{"Name": "Eugene", "Age":68, "Job": "Manager"},
                         {"Name": "Gary", "Age":5, "Job": "N/A"}],index=["Employee 5", "Employee 6"])
df1 = pd.concat([df1, new_rows])
print(df1)
```

### Reading CSV/JSON Files

```python
#To read CSV file
df2 = pd.read_csv("data1.csv")
print(df2.to_string())

'''To convert from csv to json'''
#df2_json = df2.to_json(orient="records", indent=2)
#print(df2_json)

#To read JSON file
df3 = pd.read_json("data1.json")
#print(df3.to_string())
```

### Selecting from CSV/JSON files

```python
#Selection by column

print(df2["Name"]) #will only print first & last 5 rows w/o to.string()
#To print them together
print(df2[["Name", "Height", "Weight"]])

#Selection by row(s)
df2 = pd.read_csv("data1.csv", index_col="Name")
#index_col is optional, but in case you dont want the values to be numbered 
# (to search for it easier)

print(df2.loc["Pikachu"])
print(df2.loc["Charizard", ["Height", "Weight"]])
print(df2.loc["Charizard":"Blastoise", ["Weight"]])
print(df2.iloc[0:11:2, 0:3]) #1st for rows, 2nd for columns
```

### Filtering Values

```python
print(df2[df2["Weight"] < 2])
print(df2[df2["Height"] > 3])

print(df2[(df2["Type1"] == "Fire") & (df2["Type2"] == "Flying")])
```

### Aggregation

```python
#Applies to the whole data frame:
#mean
print(df2.mean(numeric_only=True)) #finds mean for columns that are numeric

#sum
print(df2.sum(numeric_only=True))

#min
print(df2.min(numeric_only=True))

#max
print(df2.max(numeric_only=True))

#count
print(df2.count())
```

```python
#To apply to a single column:
print(df2["Height"].mean())
print(df2["Height"].sum())
print(df2["Height"].min())
#etc

#Groupby function to find something(mean,min,max,etc) of a column
# by grouping that column into different subgroups
#Here, Type 1 has diff types such as Bug, Dragon, Fairy, etc
group = df2.groupby("Type1")
print(group["Height"].mean())
print(group["Weight"].mean())
```
