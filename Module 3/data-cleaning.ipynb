```python
import pandas as pd
df = pd.read_excel(r"C:\Users\karan.chawla\Downloads\Customer Call List.xlsx")
#print(df)

#to remove duplicates rows
df = df.drop_duplicates()
#print(df)

#to remove a column
df = df.drop(columns="Not_Useful_Column")

df["Last_Name"] = df["Last_Name"].str.strip("/._")
#print(df)

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-'+ x[6:10])
df["Phone_Number"] = df["Phone_Number"].str.replace({'nan--':'', 'Na--':''})
df[["Street Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2,expand=True)
df = df.replace({'Yes':'Y', 'No':'N'}) #this is not recommended, do df["(whatever)"]
df = df.fillna('')
df = df.replace('N/a', '')
df

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df["Do_Not_Contact"]

df = df.drop(columns="Address")

df = df.reset_index(drop=True) #resets index
df
```
